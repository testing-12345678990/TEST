```python
from PIL import Image

def stack_images_vertically(image1_path, image2_path, output_path):
    # Open the images
    image1 = Image.open(image1_path)
    image2 = Image.open(image2_path)

    # Ensure the images are the same width
    if image1.size[0] != image2.size[0]:
        image2 = image2.resize((image1.size[0], image2.size[1]))

    # Calculate the new height
    new_height = image1.size[1] + image2.size[1]

    # Create a new image with the correct size
    new_image = Image.new('RGB', (image1.size[0], new_height))

    # Paste the images onto the new image
    new_image.paste(image1, (0, 0))
    new_image.paste(image2, (0, image1.size[1]))

    # Save the new image
    new_image.save(output_path)

```

```python

# Example usage:
stack_images_vertically('1.png', '2.png', 'output.png')
```

```python
from PIL import Image

def stack_small_stretch(image1_path, image2_path, output_path):
    # Open the images
    image1 = Image.open(image1_path)
    image2 = Image.open(image2_path)

    # Determine the maximum width
    max_width = max(image1.size[0], image2.size[0])

    # Resize the images if necessary
    if image1.size[0] != max_width:
        image1 = image1.resize((max_width, image1.size[1]))
    if image2.size[0] != max_width:
        image2 = image2.resize((max_width, image2.size[1]))

    # Calculate the new height
    new_height = image1.size[1] + image2.size[1]

    # Create a new image with the correct size
    new_image = Image.new('RGB', (max_width, new_height))

    # Paste the images onto the new image
    new_image.paste(image1, (0, 0))
    new_image.paste(image2, (0, image1.size[1]))

    # Save the new image
    new_image.save(output_path)

```

```python
# Example usage:
stack_small_stretch('1.png', '2.png', 'output.png')
```

```python
from PIL import Image
def stack_small_center_white(image1_path, image2_path, output_path):
    # Open the images
    image1 = Image.open(image1_path)
    image2 = Image.open(image2_path)

    # Determine the maximum width
    max_width = max(image1.size[0], image2.size[0])

    # Calculate the new height
    new_height = image1.size[1] + image2.size[1]

    # Create a new image with the correct size
    new_image = Image.new('RGB', (max_width, new_height), (255, 255, 255))  # Create a white background

    # Paste the images onto the new image
    if image1.size[0] < max_width:
        x_offset = (max_width - image1.size[0]) // 2  # Calculate the x offset to center the image
        new_image.paste(image1, (x_offset, 0))
    else:
        new_image.paste(image1, (0, 0))

    if image2.size[0] < max_width:
        x_offset = (max_width - image2.size[0]) // 2  # Calculate the x offset to center the image
        new_image.paste(image2, (x_offset, image1.size[1]))
    else:
        new_image.paste(image2, (0, image1.size[1]))

    # Save the new image
    new_image.save(output_path)


```

```python
# Example usage:
stack_small_center_white('1.png', '2.png', 'output.png')
```
